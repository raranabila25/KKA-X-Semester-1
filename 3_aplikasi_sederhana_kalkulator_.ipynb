```python
panjang = int(input("Masukkan panjang dari persegi panjang (cm): "))
lebar = int(input("Masukkan lebar dari persegi panjang (cm): "))

luas = panjang * lebar

print(f"\nLuas dari persegi panjang yaitu {luas} cm^2")
```
